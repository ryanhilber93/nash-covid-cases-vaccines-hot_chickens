```python
count_zip = pd.read_csv('COVID_CountByZipPerDate 03292021.csv')
count_zip.head(5)
```

```python
count_groupby = count_zip.drop(['Specimen Collection Date'], axis=1).groupby('Zip').sum('Cases').reset_index().sort_values(by=['Cases'], ascending=False)

#count_groupby = count_zip.drop(['Specimen Collection Date'], axis=1).sum('Cases').groupby('Zip').reset_index().sort_values(by=['Cases'], ascending=True)
count_groupby
```

```python
vaccine_protocol = pd.read_csv('vaccine_with_protocol.csv')
vaccine_protocol.head(5)
```

```python
vaccine_protocol.tail(5)
```

```python
vaccine_protocol.info()
```

```python
vaccine_protocol.rename(columns={'zipcode':"Zip"}, inplace=True)
vaccine_protocol.head()
```

```python
vaccine_filter = vaccine_protocol[vaccine_protocol.dose_nbr == 1.0]
vaccine_filter
```

```python
vaccine_group = vaccine_filter.drop(['vacc_date'], axis=1).groupby('Zip').sum().reset_index()
vaccine_group
```

```python
case_vaccine = count_groupby.merge(vaccine_group, on=["Zip"])
case_vaccine
```
